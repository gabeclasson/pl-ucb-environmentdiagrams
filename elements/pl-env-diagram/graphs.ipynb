```python
import matplotlib.pyplot as plt
```

```python
labels = ['1\n(Strongly Disagree)', '2', '3', '4', '5\n(Strongly Agree)']
def graph(question, counts, max):
    plt.bar(labels, counts)
    ax = plt.gca()
    ax.set_ylim([0, max])
    plt.title(f'Agreement with: "{question}"')
    plt.xlabel("Degree of agreement with statement")
    plt.ylabel("Number of respondents")
    plt.yticks(range(int(max) + 1))
    plt.show()

graph("I would prefer to use PrairieLearn environment\n diagrams compared to traditional pen and paper on exams.", [0, 1, 1, 4, 4], 8.5)
graph("PrairieLearn allowed me to efficiently fix\n mistakes as opposed to pen and paper.", [0, 0, 1, 1, 8], 8.5)
graph("I would recommend other courses with environment\n diagrams to use the PrairieLearn system rather than pen and paper.", [0, 1, 0, 3, 1], 3.5)
graph("Using the PrairieLearn system would allow me\n to give students ample study resources for environment diagrams.", [0, 0, 0, 2, 3], 3.5)
```
